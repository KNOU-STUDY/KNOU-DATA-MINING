```r
dout <- rep(0, 15)
dout2 <- rep(0, 15)
for(i in 1:11){
    t3 <- quantile(wine[,i], 0.75)
    t1 <- quantile(wine[,i], 0.25)
    tq <- IQR(wine[,i], 0.75)
    dout[i] <- t3 + 1.5 * tq
    dout2[i] <- t1 - 1.5 * tq
}
```

```r
nrow(wine)
outindex <- matrix(0, nrow(wine), 12)
```

```r
for (i in 1:nrow(wine)){
    for (j in 1:11){
        if(wine[i, j] > dout[j] || wine[i, j] < dout2[j]){
            outindex[i, j] <- 1
        }
    }
}
```

```r
wine2 <- wine[apply(outindex ,1 ,sum) == 0,]
```

```r
table(wine2$quality)
```

```r
head(wine2)
nrow(wine2)
```

```r
dvar <- 1:4
```

```r
prod2 <- dummy(x=prod[,dvar])
head(prod2)
```

```r
prod2 <- prod2[, -c(5, 7, 13, 25)]
```

```r
head(prod2)
```

```r
nrow(prod)
ncol(prod)
prod <- cbind(prod[, -dvar], prod2)
```

```r
nrow(prod)
ncol(prod)
```

```r
for(i in 1:ncol(prod)){
    if(!is.numeric(prod[,i])){
        prod[,i] <- as.numeric(prod[, i])
    }
}
```
